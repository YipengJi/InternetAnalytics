```python
#create a dataframe to deal with it
scores_DF=scores_data.toDF(['movieId','relevance','tagId'])
```

```python
scores_grouped=scores_DF.groupBy('movieId').agg(functions.collect_list('relevance'),functions.collect_list('tagId'))
```

```python
scores_grouped.show(3)
```

```python
import pandas
```

```python
scores=scores_grouped.collect()
```

```python
movie_matrix = np.zeros((len(tag2name),len(scores)))
movieId_matrix =[]
movie_relevances=[]

for index,row in enumerate(scores):
    movieId = row[0]
    relevances=row[1]
    tags=row[2]
    
    movieId_matrix.append(movieId)
    
    for relevance, tag in zip(relevances,tags):
        movie_matrix[tag-1,index]=relevance
        movie_relevances.append(relevance)   
    
```

```python
movie_matrix
```

```python
M=movie_matrix
```

```python
#plot that shit
```

```python
plt.boxplot(movie_relevances,0,'')
plt.title("Variance of movie relevances")
plt.ylabel("relevance")
plt.show()
```

We can see that the mean value of the relevances is close to 0.05 which is quite low. The variance shows that even if there are a few higher relevances most of the relevances are close and a little below 0.05. 
The relevance denotes the relevance of the tagg? so the tag relevances seem to be very low

```python
M_var=np.var(M,axis=1)
plt.scatter(M_var,range(1,len(M_var)+1),facecolor='skyblue')
plt.title("Variance of each tag .. ?what exactly")
plt.ylabel("tag?")
plt.xlabel("variance")
plt.show()
```
